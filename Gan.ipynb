# Hyperparameters

```python
image_dir  = 'archive'
image_size = 64
batch_size = 32
stat       = (0.5,)*3,(0.5,)*3
stat
```

# DataLoader & ImageFolder & transform

```python
from torchvision.transforms      import ToTensor
from torchvision.datasets        import ImageFolder
from torch.utils.data.dataloader import DataLoader
import torchvision.transforms as tt


transformer = tt.Compose([
    tt.Resize(image_size),
    tt.CenterCrop(image_size),
    tt.ToTensor(),
    tt.Normalize(*stat)
    
])

dateset = ImageFolder(image_dir, transform=transformer)
```

```python
train_dl = DataLoader(dateset , batch_size,shuffle=True, num_workers=0,pin_memory=True)
print(len(train_dl))
```

# Model

```python
import torch.nn as nn
import torch.nn.functional as F
```

## discrimimator

```python
#  (old_width+2*padding-kernal_size)/stride +1

discrimimator = nn.Sequential(  
    # 3*64*64
    
    nn.Conv2d(3,64,4,stride=2,padding=1,bias=False),
    nn.BatchNorm2d(64),
    nn.LeakyReLU(0.2,inplace=True),
    # 64*32*32
    
    nn.Conv2d(64,128,4,stride=2,padding=1,bias=False),
    nn.BatchNorm2d(128),
    nn.LeakyReLU(0.2,inplace=True),
    # 128*16*16
    
    nn.Conv2d(128,256,4,stride=2,padding=1,bias=False),
    nn.BatchNorm2d(256),
    nn.LeakyReLU(0.2,inplace=True),
    # 128*8*8
    
    nn.Conv2d(256,512,4,stride=2,padding=1,bias=False),
    nn.BatchNorm2d(512),
    nn.LeakyReLU(0.2,inplace=True),
    # 512*4*4
    
    nn.Conv2d(512,1,4,stride=1,padding=0,bias=False),
    # 1*1*1
    
    
    nn.Flatten(),
    nn.Sigmoid()
).cuda()
```

```python
from torchviz import make_dot
img, _ = dateset[0]
y = discrimimator(img.unsqueeze(0).cuda())
make_dot(y.mean(), params=dict(discrimimator.named_parameters()), show_attrs=False, show_saved=False).render(
    "discrimimator", format="svg")
```

## generator

```python
latent_size = 128
```

```python
# (old_width - 1)*stride - 2*padding + kernal_size

generator = nn.Sequential(
    
    # latent_size * 1 * 1
    
    nn.ConvTranspose2d(latent_size,512, kernel_size=4,stride=1,padding=0,bias=False),
    nn.BatchNorm2d(512),
    nn.ReLU(True),   
    # 512 * 4 * 4
    
    nn.ConvTranspose2d(512,256, kernel_size=4,stride=2,padding=1,bias=False),
    nn.BatchNorm2d(256),
    nn.ReLU(True),   
    # 256 * 8 * 8  ( (4-1)*2 - 2*1 + 4 )
    
    nn.ConvTranspose2d(256,128, kernel_size=4,stride=2,padding=1,bias=False),
    nn.BatchNorm2d(128),
    nn.ReLU(True),   
    # 128 * 16 * 16
    
    nn.ConvTranspose2d(128,64, kernel_size=4,stride=2,padding=1,bias=False),
    nn.BatchNorm2d(64),
    nn.ReLU(True),   
    # 64 * 32 * 32
    
    nn.ConvTranspose2d(64,3, kernel_size=4,stride=2,padding=1,bias=False),
    nn.Tanh()
    # 3* 32 * 32
).cuda()
```

# Train it

```python
def train_discrimimator(real_image, opt_d):
    # clear discrimimator gradients
    opt_d.zero_grad()
    
    # past real image through discrimimator
    real_preds  = discrimimator(real_image)
    real_target = torch.ones(real_image.size(0),1).cuda()
    real_loss   = F.binary_cross_entropy(real_preds, real_target)
    real_score  = torch.mean(real_preds).item()
    
    # Generate fake images
    xb         = torch.randn(batch_size, latent_size, 1, 1).cuda()
    fake_image = generator(xb)
    
    # past fake image through discrimimator
    fake_preds  = discrimimator(fake_image)
    fake_target = torch.zeros(fake_image.size(0),1).cuda()
    fake_loss   = F.binary_cross_entropy(fake_preds, fake_target)
    fake_score  = torch.mean(fake_preds).item()
    
    # update discrimimator weight
    loss  = real_loss + fake_loss
    loss.backward()
    opt_d.step()
    
    return loss.item(), real_score, fake_score
    

def train_generator(opt_g):
    # clear generator gradients
    opt_g.zero_grad()
    
    # Generate fake images
    xb         = torch.randn(batch_size, latent_size, 1, 1).cuda()
    fake_image = generator(xb)
    
    # try to fool the discrimimator
    preds = discrimimator(fake_image)
    target = torch.ones(preds.size(0),1).cuda()
    loss   = F.binary_cross_entropy(preds, target)
    
    loss.backward()
    opt_g.step()
    return loss.item()
    
    
    
    
    
```
